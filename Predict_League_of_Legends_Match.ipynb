# Prepare Predictor Variables

1. Extract final value from the golddiff value

```python
# Gold difference is evaluated as goldblue - goldred
# Negative gold difference implies that red team had a bigger gold value near the end of the match
matches['final_golddiff'] = matches['golddiff'].apply(lambda x : eval(x)[-1])
matches['final_golddiff']
```

```python
matches['gamelength']
```

```python
# Counting number of kills of blue team
matches['bKills_count'] = matches['bKills'].apply(lambda x : len(eval(x)) if x else 0)
matches['bKills_count']
```

```python
# Counting number of kills of red team
matches['rKills_count'] = matches['rKills'].apply(lambda x : len(eval(x)) if x else 0)
matches['rKills_count']
```

```python
# Evaluate K/D ratio for blue team
matches['bKD_ratio'] = matches['bKills_count'] / matches['rKills_count']
matches['bKD_ratio']
```

```python
# Evaluate K/D ratio for red team
matches['rKD_ratio'] = matches['rKills_count'] / matches['bKills_count']
matches['rKD_ratio']
```

```python
# Counting number of towers of blue team
matches['bTowers_count'] = matches['bTowers'].apply(lambda x : len(eval(x)) if x else 0)
matches['bTowers_count']
```

```python
# Counting number of towers of red team
matches['rTowers_count'] = matches['rTowers'].apply(lambda x : len(eval(x)) if x else 0)
matches['rTowers_count']
```

```python
# Counting number of inhibitors of blue team
matches['bInhibs_count'] = matches['bInhibs'].apply(lambda x : len(eval(x)) if x else 0)
matches['bInhibs_count']
```

```python
# Counting number of inhibitors of red team
matches['rInhibs_count'] = matches['rInhibs'].apply(lambda x : len(eval(x)) if x else 0)
matches['rInhibs_count']
```

```python
# Counting number of dragons of blue team
matches['bDragons_count'] = matches['bDragons'].apply(lambda x : len(eval(x)) if x else 0)
matches['bDragons_count']
```

```python
# Counting number of dragons of red team
matches['rDragons_count'] = matches['rDragons'].apply(lambda x : len(eval(x)) if x else 0)
matches['rDragons_count']
```

```python
# Counting number of barons of blue team
matches['bBarons_count'] = matches['bBarons'].apply(lambda x : len(eval(x)) if x else 0)
matches['bBarons_count']
```

```python
# Counting number of barons of red team
matches['rBarons_count'] = matches['rBarons'].apply(lambda x : len(eval(x)) if x else 0)
matches['rBarons_count']
```

```python
# Counting number of heralds of blue team
matches['bHeralds_count'] = matches['bHeralds'].apply(lambda x : len(eval(x)) if x else 0)
matches['bHeralds_count']
```

```python
# Counting number of heralds of red team
matches['rHeralds_count'] = matches['rHeralds'].apply(lambda x : len(eval(x)) if  x else 0)
matches['rHeralds_count']
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
train = matches[matches["Year"] < 2016]
test = matches[matches["Year"] >= 2016]
predictors = ["final_golddiff", "gamelength", "bKills_count", "rKills_count", "bKD_ratio", "rKD_ratio", "bTowers_count", "rTowers_count", "bInhibs_count", "rInhibs_count", "bDragons_count", "rDragons_count", "bBarons_count", "rBarons_count", "bHeralds_count", "rHeralds_count"]
```

```python
rf.fit(train[predictors], train["target_encoded"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
error = accuracy_score(test["target_encoded"], preds)
```

```python
blue_data = matches.rename(columns={
    'League': 'League',
    'Season': 'Season',
    'Year': 'Year',
    'blueTeamTag': 'Team',
    'redTeamTag': 'Opponent',
    'bResult': 'Result',
    'gamelength': 'Length',
    'final_goldblue': 'Team Gold',
    'final_goldred': 'Opponent Gold',
    'bKills_count': 'Kills',
    'bKD_ratio': 'KD Ratio',
    'bTowers_count': 'Towers',
    'bInhibs_count': 'Inhibs',
    'bDragons_count': 'Dragons',
    'bBarons_count': 'Barons',
    'bHeralds_count': 'Heralds'
})[['League', 'Season', 'Year', 'Team', 'Opponent', 'Result', 'Length', 'Kills', 'KD Ratio', 'Towers', 'Inhibs', 'Dragons', 'Barons', 'Heralds', 'Team Gold', 'Opponent Gold']]
blue_data
```

```python
# Calculate final gold difference for blue teams
blue_data['Gold Difference'] = blue_data['Team Gold'] - blue_data['Opponent Gold']
blue_data
```

```python
red_data = matches.rename(columns={
    'League': 'League',
    'Season': 'Season',
    'Year': 'Year',
    'redTeamTag': 'Team',
    'blueTeamTag': 'Opponent',
    'rResult': 'Result',
    'gamelength': 'Length',
    'final_goldred': 'Team Gold',
    'final_goldblue': 'Opponent Gold',
    'rKills_count': 'Kills',
    'rKD_ratio': 'KD Ratio',
    'rTowers_count': 'Towers',
    'rInhibs_count': 'Inhibs',
    'rDragons_count': 'Dragons',
    'rBarons_count': 'Barons',
    'rHeralds_count': 'Heralds'
})[['League', 'Season', 'Year', 'Team', 'Opponent', 'Result', 'Length', 'Kills', 'KD Ratio', 'Towers', 'Inhibs', 'Dragons', 'Barons', 'Heralds', 'Team Gold', 'Opponent Gold']]
red_data
```

```python
# Calculate final gold difference for red teams
red_data['Gold Difference'] = red_data['Team Gold'] - red_data['Opponent Gold']
red_data
```

```python
# Combine both into a single dataframe
team_data = pd.concat([blue_data, red_data], ignore_index=True)
team_data
```

```python
grouped_matches = team_data.groupby('Team')
```

```python
group = grouped_matches.get_group("kt").sort_values("Year")
group
```

```python
def rolling_averages(group, cols, new_cols):
  group = group.sort_values("Year")
  rolling_stats = group[cols].rolling(3, closed='left').mean()
  group[new_cols] = rolling_stats
  group = group.dropna(subset=new_cols)
  return group
```

```python
cols = ["Length", "Kills", "KD Ratio", "Towers", "Inhibs", "Dragons", "Barons", "Heralds", "Team Gold", "Gold Difference"]
new_cols = [f"{c}_rolling" for c in cols]

rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = team_data.groupby("Team").apply(lambda x: rolling_averages(x, cols, new_cols))
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel('Team')
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
# Prepare new predictor variables
predictors = ["Length", "Kills", "KD Ratio", "Towers", "Inhibs", "Dragons", "Barons", "Heralds", "Team Gold", "Gold Difference"]
```

```python
def make_predictions(data, predictors):
    train = data[data["Year"] < 2016]
    test = data[data["Year"] >= 2016]
    rf.fit(train[predictors], train["Result"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["Result"], predicted=preds), index=test.index)
    error = precision_score(test["Result"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)
```

# Combining Home and Away Predictions

> Note: The dataset computed in this section might present some inaccuracies due to the duplicate data.

```python
class MissingDict(dict):
    __missing__ = lambda self, key: key

map_value = {
    "kt": "KT",
    "yoeFW": "FW",
    "ahq": "AHQ",
    "HKEs": "HKE",
    "CJ": "CJE",
    "as": "AS",
    "WS": "NWS",
    "g3x": "G3X",
    "Mac": "MAC",
    "DoR": "DOR",
    "bbq": "BBQ"
}

mapping = MissingDict(**map_value)
```

```python
combined["new_team"] = combined["Team"].map(mapping)
```

```python
combined["new_opponent"] = combined["Opponent"].map(mapping)
```

```python
combined
```

```python
merged = combined.merge(combined, left_on=["Year", "League", "Season", "Length", "new_team", "new_opponent"], right_on=["Year", "League", "Season", "Length", "new_opponent", "new_team"])
```
